### Load packages

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

### Load sample image

```python
img = cv2.imread('./img/1.jpg', 0) #Load in grayscale
```

```python
plt.figure(figsize=(10,10))
plt.imshow(img, cmap='gray')
```

### Bi-modal segementation

Bi-modal segementation extracts data from noisy image. The redline in the histogram below shows the threshold where filters noise

```python
ret, imgf = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

plt.figure()
plt.hist(img.ravel(), 256)
plt.axvline(x=ret, color='r', linestyle='dashed', linewidth=2)
plt.title('Histogram'), plt.xticks([]), plt.yticks([])
```

```python
plt.figure(figsize=(10,10))
plt.imshow(imgf,cmap = 'gray') # imgf == image filtered 
plt.title('Otsu thresholding'), plt.xticks([]), plt.yticks([])
plt.show()
```
